# Abstract

3D printing objects comes with the disadvantage of having to print extra 'support structures' so that the printed object holds its shape. These lead to disfigurations during post-processing and thus should be minimized. We can minimize them printing the objects in parts and gluing them later. In this exercise, we try to find the plane parallel to to the yz plane and passing through the X axis along which we can cut the object in order to lead to the fewest support structures.

# Introduction

*Fused Deposition Modelling (FDM)* refers to a 3D printing technique in which material such as plastic or metal is unwound from a coil and deposited by a heated nozzle in layers on a surface to produce a 3D printout resembling a 3D model (usually provided in .STL file format). It is an *additive* manufacturing technique **[1]**, which means that the amount of material used to make an object is very close to the amount of material it actually consists of. This saves wastages and thus the cost of printing items. This technique is widely used to produce prototypes rapidly which can then be produced by other manufacturing methods. However, the technique does come with some disadvantages. This technique is not quite scalable yet. Depending on the fineness of the printer, the output might not be very smooth. Also, since the material is molten, it might start to bend while being printed and thus need 'support structures' which can cause scars on the model while being removed during post processing. This second problem is what we address in this exercise.

This happens because in such models, there is an increased amount of overhang in each subsequent printed layer which causes drooping. This problem is very hard to avoid with current technology and viable materials. One proposed solution is having 'support structures like the ones in *Figure 2*. These cause problems since they not only increase wastages, they lead to scars on the output when they are removed. Thus, we want to minimize them. One way to do this is cutting the object along certain planes and glue these parts together after printing. In this project, we try to find the ideal plane parallel to the YZ plane and passing through the X axis to cut a given object using genetic algorithms. We generate a population of candidate points on the X axis the plane may pass through, and then evolve this population using *DEAP*. we produce the plane passing though the point in X axis which leads to the least amount of support structures.

# Methodology

The required plane along which the given figure should be cut can be parameterized with a single number- the point along the X axis through which it passes. We used *DEAP* in order to find the optimal value of this number.

*DEAP* initially produces a population of size 15 of X values (floating point), ranging from 0 to 100. Then, we compute the cost associated with each of this individuals, which is just the value returned by the slicing program. This cost represents the amount of support structures required to produce the output and obviously needs to be minimized. This represents the amount of support structures needed while printing the two parts of the output. Our goal is to minimize this value, which is why the weight is set to -1 in the FitnessMin object.

We then use the built in *DEAP* to mutate and mate the individuals from this generation to produce the next generation. We then repeat this process and choose the best individual that we come across. We use blend for mating and Guass for mutation.

Here are the parameters that we used while carrying out our implementation of the genetic algorithm:

* Number of Generations: 20
* Population size: 15
* Crossover probability: 0.8
* Mutation probability: 0.1

We keep track of the following statistics for each generation:

* Minimum cost (ie. minimum support structures)
* Maximum cost (ie. maximum support structures)
* The average cost in each generation
* Standard Deviation of cost values in each generation

As it turned out, we did not have to change the parameter values very much. We were able to find and confirm the optimum cutting plane very quickly, as discussed in the next section.
